```python
import numpy as np
from collections import defaultdict
from itertools import combinations
from typing import List, Tuple, Dict, Any

class ExteriorPolynomial:
    """
    Represents polynomials in the exterior algebra Z_2[x_1,...,x_p]/(x_i^2).
    A polynomial is stored as a dictionary mapping monomial tuples to coefficients { (0,1):1, (2,3):1 } for x_1*x_2 + x_3*x_4.
    """
    def __init__(self, expr_dict: Dict[Tuple, int], num_vars: int):
        self.num_vars = num_vars
        self.terms = {k: v % 2 for k, v in expr_dict.items() if v % 2 != 0}
    
    def __add__(self, other):
        result_terms = self.terms.copy()
        for monomial, coeff in other.terms.items():
            current_coeff = result_terms.get(monomial, 0)
            new_coeff = (current_coeff + coeff) % 2
            if new_coeff == 1:
                result_terms[monomial] = 1
            elif monomial in result_terms:
                del result_terms[monomial]
        return ExteriorPolynomial(result_terms, self.num_vars)
    
    def __mul__(self, other):
        result_terms = defaultdict(int)
        for mon1, coeff1 in self.terms.items():
            for mon2, coeff2 in other.terms.items():
                if set(mon1).intersection(set(mon2)):
                    continue
                
                new_monomial = tuple(sorted(set(mon1) | set(mon2)))
                new_coeff = (coeff1 * coeff2) % 2
                result_terms[new_monomial] = (result_terms[new_monomial] + new_coeff) % 2
        
        final_terms = {k: v for k, v in result_terms.items() if v != 0}
        return ExteriorPolynomial(final_terms, self.num_vars)

    def __len__(self):
        """Return the number of terms in the polynomial."""
        return len(self.terms)

    def degree(self):
        if not self.terms:
            return -1
        return max(len(monomial) for monomial in self.terms.keys())
    
    def is_zero(self):
        return not self.terms

    def __str__(self):
        if self.is_zero():
            return "0"
        
        sorted_terms = sorted(self.terms.keys(), key=lambda m: (len(m), m))
        
        result = []
        for monomial in sorted_terms:
            if not monomial:
                result.append("1")
            else:
                result.append("*".join(f"x_{i+1}" for i in monomial))
        return " + ".join(result)
    
    def __repr__(self):
        return f"ExteriorPolynomial({self})'"

def create_monomial(variables: Tuple[int], num_vars: int) -> ExteriorPolynomial:
    return ExteriorPolynomial({tuple(sorted(variables)): 1}, num_vars)

def elementary_symmetric_polynomial(k: int, variable_indices: List[int], num_vars: int) -> ExteriorPolynomial:
    if k == 0:
        return ExteriorPolynomial({(): 1}, num_vars)
    if k < 0 or k > len(variable_indices):
        return ExteriorPolynomial({}, num_vars)
    
    result = ExteriorPolynomial({}, num_vars)
    for subset in combinations(variable_indices, k):
        result = result + create_monomial(subset, num_vars)
    return result

def get_monomials_of_degree(degree: int, variable_indices: List[int], num_vars: int) -> List[ExteriorPolynomial]:
    return [create_monomial(subset, num_vars) for subset in combinations(variable_indices, degree)]

def polynomial_to_vector(poly: ExteriorPolynomial, monomial_basis: List[ExteriorPolynomial]) -> np.ndarray:
    vector = np.zeros(len(monomial_basis), dtype=int)
    mono_map = {list(m.terms.keys())[0]: i for i, m in enumerate(monomial_basis) if m.terms}
    for mono_tuple in poly.terms:
        if mono_tuple in mono_map:
            vector[mono_map[mono_tuple]] = 1
    return vector

def vector_to_polynomial(vector: np.ndarray, monomial_basis: List[ExteriorPolynomial], num_vars: int) -> ExteriorPolynomial:
    result = ExteriorPolynomial({}, num_vars)
    for i, coeff in enumerate(vector):
        if coeff % 2 != 0:
            result = result + monomial_basis[i]
    return result

def null_space_z2(matrix):
    A = matrix.astype(int) % 2
    m, n = A.shape
    if m == 0: return np.eye(n, dtype=int)
    
    rank = 0
    pivot_cols = []
    for col in range(n):
        if rank < m and A[rank, col] == 1:
            pivot_cols.append(col)
            for row in range(m):
                if row != rank and A[row, col] == 1: A[row] ^= A[rank]
            rank += 1
        else:
            found_pivot = False
            for r in range(rank + 1, m):
                if A[r, col] == 1:
                    A[[rank, r]] = A[[r, rank]]
                    pivot_cols.append(col)
                    for row in range(m):
                        if row != rank and A[row, col] == 1: A[row] ^= A[rank]
                    rank += 1
                    found_pivot = True
                    break
            if not found_pivot: continue

    free_vars = [i for i in range(n) if i not in pivot_cols]    
    null_basis = []
    for free_var in free_vars:
        solution = np.zeros(n, dtype=int)
        solution[free_var] = 1
        for i, pivot_col in reversed(list(enumerate(pivot_cols))):
            solution[pivot_col] = sum(A[i, j] * solution[j] for j in free_vars) % 2
    
        null_basis.append(solution)
        
    return np.array(null_basis, dtype=int) if null_basis else np.zeros((0, n), dtype=int)

def compute_annihilator_space(target: ExteriorPolynomial, variable_indices: List[int], 
                            degree: int, num_vars: int) -> List[ExteriorPolynomial]:
    candidate_monomials = get_monomials_of_degree(degree, variable_indices, num_vars)
    if not candidate_monomials: return []

    product_degree = degree + target.degree()
    all_product_monomials = []
    for d in range(product_degree + 1):
        all_product_monomials.extend(get_monomials_of_degree(d, variable_indices, num_vars))

    constraint_matrix = []
    for prod_mono in all_product_monomials:
        if prod_mono.is_zero(): continue
        prod_mono_tuple = list(prod_mono.terms.keys())[0]
        row = [p.terms.get(prod_mono_tuple, 0) for p in [cand * target for cand in candidate_monomials]]
        if any(row):
            constraint_matrix.append(row)

    if not constraint_matrix: return candidate_monomials
    A = np.array(constraint_matrix, dtype=int) % 2
    
    null_space_vectors = null_space_z2(A)
    return [vector_to_polynomial(v, candidate_monomials, num_vars) for v in null_space_vectors]

def row_basis_z2(matrix):
    if matrix.size == 0: return matrix
    A = matrix.astype(int) % 2
    m, n = A.shape
    rank = 0
    for col in range(n):
        if rank < m:
            pivot_row = rank
            while pivot_row < m and A[pivot_row, col] == 0: pivot_row += 1
            if pivot_row < m:
                A[[rank, pivot_row]] = A[[pivot_row, rank]]
                for row in range(m):
                    if row != rank and A[row, col] == 1: A[row] ^= A[rank]
                rank += 1
    return A[:rank]

def matrix_rank_z2(matrix):
    return row_basis_z2(matrix.copy()).shape[0]

def compute_quotient_basis(space: List[ExteriorPolynomial], ideal_generators: List[ExteriorPolynomial], 
                         num_vars: int) -> List[ExteriorPolynomial]:
    if not space: return []
    if not ideal_generators: return space

    degree = space[0].degree()
    all_vars = set()
    for poly in space + ideal_generators:
        for mono in poly.terms: all_vars.update(mono)
    var_list = sorted(list(all_vars))
    monomial_basis = get_monomials_of_degree(degree, var_list, num_vars)

    if not monomial_basis: return []
        
    space_matrix = np.array([polynomial_to_vector(p, monomial_basis) for p in space])
    ideal_generators_degree = [p for p in ideal_generators if p.degree() == degree and not p.is_zero()]
    
    if not ideal_generators_degree: return space

    ideal_matrix = np.array([polynomial_to_vector(p, monomial_basis) for p in ideal_generators_degree])
    ideal_basis_matrix = row_basis_z2(ideal_matrix)
    ideal_rank = ideal_basis_matrix.shape[0]

    quotient_basis = []
    current_basis_matrix = ideal_basis_matrix.copy()
    for i, p_vec in enumerate(space_matrix):
        test_matrix = np.vstack([current_basis_matrix, p_vec.reshape(1, -1)]) if current_basis_matrix.size > 0 else p_vec.reshape(1,-1)
        if matrix_rank_z2(test_matrix) > current_basis_matrix.shape[0]:
            quotient_basis.append(space[i])
            current_basis_matrix = row_basis_z2(test_matrix)
            
    return quotient_basis

def apply_variable_mapping(poly: ExteriorPolynomial, mapping: Dict[int, int], num_vars: int) -> ExteriorPolynomial:
    new_terms = defaultdict(int)
    for monomial, coeff in poly.terms.items():
        new_monomial_vars = [mapping.get(var, var) for var in monomial]
        new_monomial = tuple(sorted(new_monomial_vars))
        new_terms[new_monomial] = (new_terms[new_monomial] + coeff) % 2
    return ExteriorPolynomial({k: v for k, v in new_terms.items() if v != 0}, num_vars)

def sparsify_basis(basis: List[ExteriorPolynomial]) -> List[ExteriorPolynomial]:
    """
    Finds a sparser basis for the given vector space using a greedy reduction algorithm.
    """
    if not basis:
        return []

    sparse_basis = basis[:]
    num_vectors = len(sparse_basis)

    changed_in_iteration = True
    while changed_in_iteration:
        changed_in_iteration = False
        sparse_basis.sort(key=len, reverse=True) # Process larger polynomials first

        for i in range(num_vectors):
            for j in range(num_vectors):
                if i == j: continue
                
                sum_poly = sparse_basis[i] + sparse_basis[j]
                
                if len(sum_poly) < len(sparse_basis[i]):
                    sparse_basis[i] = sum_poly
                    changed_in_iteration = True
    
    sparse_basis.sort(key=len)
    return sparse_basis


def find_primitive_generators(p: int, q: int, w: int) -> List[Tuple[List[int], ExteriorPolynomial]]:
    """
    Main algorithm to find primitive generators, optimized for sparsest basis.
    """
    print(f"Finding primitive generators for ({p},{q},{w}) code (optimized for minimal terms)")
    
    k = q - w
    all_primitive_generators = {}
    
    for t in range(q + 1, p + 1):
        print(f"\n--- Analyzing support size t = {t} ---")
        
        I_t_canonical = list(range(t))
        target_polynomial = elementary_symmetric_polynomial(k, I_t_canonical, p)
        
        print(f"Target elementary symmetric polynomial S_{k} on support {{x_1,...,x_{t}}}:")
        print(f"  > {target_polynomial}")
        
        annihilator_degree = t - q
        annihilator_space = compute_annihilator_space(target_polynomial, I_t_canonical, annihilator_degree, p)
        
        if not annihilator_space:
            print(f"No annihilators found at level t = {t}")
            continue

        redundant_generators = []
        for t_prime in range(q + 1, t):
            if t_prime in all_primitive_generators:
                for I_t_prime_canonical, A_prime in all_primitive_generators[t_prime]:
                    for embedding in combinations(I_t_canonical, t_prime):
                        var_mapping = {I_t_prime_canonical[i]: embedding[i] for i in range(t_prime)}
                        A_embedded = apply_variable_mapping(A_prime, var_mapping, p)
                        monomial_degree = t - t_prime
                        if monomial_degree == 0:
                            redundant_generators.append(A_embedded)
                        else:
                            remaining_vars = [v for v in I_t_canonical if v not in embedding]
                            for monomial_vars in combinations(remaining_vars, monomial_degree):
                                monomial = create_monomial(monomial_vars, p)
                                redundant_generators.append(A_embedded * monomial)

        raw_primitive_generators = compute_quotient_basis(annihilator_space, redundant_generators, p)
        
        if raw_primitive_generators:
            print(f"Found {len(raw_primitive_generators)} primitive generators at t = {t}. Applying sparsification...")
            
            sparse_generators = sparsify_basis(raw_primitive_generators)
            
            all_primitive_generators[t] = [(I_t_canonical, gen) for gen in sparse_generators]
            
            print(f"  Final basis for t = {t}:")
            for i, gen in enumerate(sparse_generators):
                print(f"    Generator {i+1} (terms: {len(gen)}): {gen}")
        else:
            print(f"No primitive generators at level t = {t}")
    
    final_result = []
    for t in sorted(all_primitive_generators.keys()):
        final_result.extend(all_primitive_generators[t])
    
    return final_result


if __name__ == "__main__":
    print("="*50)
    print("=== Testing (6,3,1) Bootstrap Product Code ===")
    primitive_gens = find_primitive_generators(6, 3, 1)
    
    print(f"\n=== Final Results for (6,3,1) ===")
    print(f"Found {len(primitive_gens)} primitive generators (up to symmetry):")
    for i, (support, generator) in enumerate(primitive_gens):
        print(f"Generator {i+1}:")
        print(f"  Support: {[x+1 for x in support]} (variables x_{support[0]+1}, ..., x_{support[-1]+1})")
        print(f"  Polynomial (terms: {len(generator)}): {generator}")
```
